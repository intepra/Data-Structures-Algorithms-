Write a recursive function to compute the Fibonacci sequence. How does the performance of the recursive function compare to that of an iterative version?

```python
# I will be comparing the performance of the recursive and iterative versions by the magic cell %timeit 
# first: recursive version 
def fib_rec(n):
    #check that the input is a positive integer 
    if type(n) != int:
        raise TypeError ("n must be a positive integer")
    if n < 0:
        raise ValueError("n must be a positive integer")
    #base case 
    if n == 0 or n == 1:
        return n 
    #Recursion
    else:
        return fib_rec(n-1) + fib_rec(n-2)
```

```python
#test my function
fib_rec(14)
```

```python
#using timeit for recursive for n = 14
%timeit fib_rec(14)
```

```python
#implementing recusrive for a range between (1,30) and then use timeit for measurement: 
for n in range(1,30):
    print(n, ":", fib_rec(n)) 
```

```python
%%timeit 
for n in range(1,30):
    print(n, ":", fib_rec(n))     
            
```

```python
#now we will find fibonacci using the iterative method 
def fib_iter(n):
    a = 0
    b = 1
    for i in range (n):
        c = a + b
        a = b
        b = c
    return a
```

```python
#testing my function
fib_iter(14)
```

```python
##using timeit for iterative for n = 14
%timeit fib_iter(14)
```

```python
#implementing iterative for a range between (1,30) and then use timeit for measurement: 
for n in range(1,30):
    print(n, ":", fib_iter(n)) 
```

```python
%%timeit
for n in range(1,30):
    print(n, ":", fib_iter(n)) 
```

The performance of the recursive function compare to that of an iterative version: we used timeit to measure both versions for a range between (1, 30). We could see that the recusrive version took 1 loop, best of 3: 2.6 s per loop and the iterative 100 loops, best of 3: 19.9 ms per loop. The recurisve one is slower than the iterative which means it uses more memory and started being slower in large cases. The iterative is longer code but faster and uses less memory than recursive. Recursive solution has O(2^n) time complexity while the iterative solution has O(n) time complexity

The postive aspect about the recurisve fib function: it took into consideration cases like string/integer and negative numbers or n < 1.
The limitation: it did not start fib sequence from one but started instead from "0".
